```python
x=df.drop('Species',axis=1)
y=df['Species']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
#Deep Learning 
#1)Build the Model
#2)Compile the Model
#3)Run the model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()
model.add(Dense(32,input_dim=4,activation='relu'))#input layer
model.add(Dense(16,activation='relu'))#hidden layer 1
model.add(Dense(8,activation='relu'))#hidden layer 2
model.add(Dense(8,activation='relu'))#hidden layer 3
model.add(Dense(3,activation='softmax'))#output layer #3types of flower[0.2,0.2,0.6]

#softmax returns the probabilities

#Output Layer Activation
#binary - sigmoid
#categorical- softmax
#regression- no activation

model.summary()  

```

```python
#step 2- compile the model
#LOSS
#CATEGORICAL=parse_categorical_crossentropy
#BINARY=binary_crossentropy
#regression=mse
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
#step 3- Run the model
model.fit(x_train,y_train,validation_split=0.2,epochs=100,verbose=1)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
y_pred=np.argmax(y_pred,axis=1)
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(y_test,y_pred)
```
