```python
text='''Apple inc is planning to open a new office in bengaluru 2025.
This will create  1000s of jobs for software engineers.
Tim cook announced  this in a press conference in california'''
```

```python
sentences=sent_tokenize(text)
print(sentences)
```

```python
word=word_tokenize(text)
print(word)
```

```python
#remove stopwords
stop_words_list=stopwords.words('english')
filtered_words=[i for i in word if i.lower() not in stop_words_list]
```

Stemming

```python
stemmer=PorterStemmer()
stemmed=[stemmer.stem(i) for i in filtered_words]
```

Lemmatization

```python
lemma=WordNetLemmatizer()
lematized=[lemma.lemmatize(i) for i in filtered_words]
```

```python
pos_tags=pos_tag(filtered_words)
print(pos_tags)
#nnp=proper noun
#jj=adjective
#cd=cardinal number
#vbg-present participle
```

```python
ne_tree=ne_chunk(pos_tags)
```
