```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read the data
df=pd.read_csv("spam_ham_dataset.csv")
```

```python
#cleaning the data,remove digits ,punctuation signs etc
#nlp-making our computer understand natural language.
#feature encoding-the natural lang english can be learnt as numbers and then generate more text using that pattern.
#inside chatgpt somekind of mathematical function are converting our human natural language and processing it and giving back us the natural answers.
df
```

```python
text=df['text'][0]

import re
import string
from nltk.stem import PorterStemmer,LancasterStemmer,SnowballStemmer
st=PorterStemmer()
#cleaning data

def clean_text(text):
    text=text.lower() #lower whole text
 
    text=re.sub(r'\d+','',text) #remove digits
    
  
    #removing punctuationS
    text=text.translate(str.maketrans('','',string.punctuation))
  
   
    #removing extra spaces from beginning and end of text
    text=text.strip()

    #target particular word in a text
    text=text.replace('subject','').strip()

    text=' '.join([st.stem(word) for word in text.split()]) 

  

    return text






```

#nltk-natural language toolkit
#pip install nltk

#lemmatization-basic english converted to root words.,base words



```python
#another way to apply function of whole dataset
df['clean_text']=df['text'].apply(clean_text)
```

```python
from  sklearn.feature_extraction.text import TfidfVectorizer
#TFIDF-Term Frequency -inverse document frequency
vectorizer=TfidfVectorizer(stop_words='english')
#stop words=[an,is,the,been,our,to,i,intoS]
tfidf=vectorizer.fit_transform(['this is a test meassage which has dublicate dublicate values'])

```

```python
pd.DataFrame(tfidf.toarray(),columns=vectorizer.get_feature_names_out())
```

```python
x=vectorizer.fit_transform(df['clean_text'])
y=df['label_num']
```

```python
from sklearn.model_selection  import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB
mn=MultinomialNB()

mn.fit(x_train,y_train)
```

```python
y_pred=mn.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
#predict a new email
def predict_email(text):
    cleaned=clean_text(text)
    vectorized=vectorizer.transform([cleaned])
    pred=mn.predict(vectorized)
    if pred==1:
        print("This is a spam email")
    else :
        print('This is a genuine email')   
        
         
```

```python
example_text='Congratulations!!!,you have won a free samsung phone,click here to claim!'
predict_email(example_text)
```
