```python
x=df.drop('gender',axis=1)
```

```python
y=df['gender']
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x=ss.fit_transform(x)

#data->mean=0 and std dev=1
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,MODEy_test=train_test_split(x,y,random_state=100,test_size=0.20)
```

```python
#types of bayes
#bernoulli naive bayes-for binary and boolean features
#multinomial naive bayes-for discrete features
#gaussion naive bayes-for continuous values
```

```python
x
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(x_train,y_train)
```

```python
y_pred_train=model.predict(x_train)
```

```python
y_pred_test=model.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_train,y_pred_train))
```

```python
print(classification_report(y_test,y_pred_test))
```
