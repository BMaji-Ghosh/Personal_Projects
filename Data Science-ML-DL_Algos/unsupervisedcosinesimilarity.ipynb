```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("bollywood1to950.csv")
```

```python
df.info()
```

```python
df.drop('Year',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.describe()
```

```python
df=df.dropna()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
MS=MinMaxScaler()
features=MS.fit_transform(df.drop('Name',axis=1))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(features)
```

```python
cosine_df=pd.DataFrame(similarity,index=df['Name'],columns=df['Name'])
```

```python
cosine_df
```

```python
def recommend_bollywood_movies(movie_name):
    print("Movie recommendation for:",movie_name)
    x=1
    for i in cosine_df['Race 3'].sort_values(ascending=False)[1:].head(5).index:
        print(f'{x} {i}')
        x+=1
```

```python
recommend_bollywood_movies('Race 3')
```
